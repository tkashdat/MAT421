# Base 2 and Base 10 Numerical Representation
- Numbers represented in the Base 10 number system are the common form.  However, numbers can be represented in many systems of representation, as long as there is a algorithm for conversion.  This is particularly useful when there are limitations in the how numbers can used and stored, as is common in computing.  

- Base 2 uses digits 0 and 1 (in contrast to 0-9 in base 10) and exponent base 2.

```python
# This function converts a Base 10 number to Base 2

def base_conversion(num1):
    binary_list = []
    while num1 > 0:
        r = num1 % 2
        num1 = num1 // 2
        binary_list.append(r)
    binary_list.reverse()
    return binary_list
```

```python
base_conversion(17)
```

## IEEE754

- Limitations of computing hardware necessitates the floating point number representation from decimal to finite, fixed binary or IEEE754
- IEEE754 allows for sufficient representation of digits (as bits) for accurate computation, which is limited in fixed point binary
- IEE754 uses a sign indicator (indicates positive or negative), a exponent (the power of the exponenetial base 2 minus offset), and a mantissa (the decimal portion or coefficient).
- Sign: 0 for postiive and 1 for negative
- Exponent: power from performing scientific notation in base 2, minus the offset or bias (1023 for 64 bit)
- Mantissa: the decimal portion from the base 2 number after performing scientific notation, 23 digits

## Round-off Error

- There is a difference between the true representation of a decimal and a floating point number, since there is ulitamtely a sacrafice in precision for computer representation  
- The difference between the true value of a base10 decimal and its computational analogue is called the round-off error
- This is readily seen when making logical comparisons between decimal calculations and expected exact decimal output since the floating point number is a rounded approximation
- Round-off errors can accumulate and have a significant effect on sequential calculations

```python
x = .6501
y = .8889

# computer representation of sum
z = x + y
print(z)

# comparison to computed answer
print(z == 1.539)

#applying round function to 3 digits before comparison
print(round(z, 3) == 1.539)
```
